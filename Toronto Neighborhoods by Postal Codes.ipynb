# Toronto Neighborhoods Analysis and Clustering

```python
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
```

Scrape Data from Wikipedia

```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'html.parser')
print(soup.prettify())
```

```python
wtable = soup.find('table',{'class':'wikitable sortable'})
wtable
```

```python
df = pd.read_html(str(wtable))[0]
df.head()
```

Clean Dataframe

```python
df1=df.replace('Not assigned', np.nan)
df1.head()
```

```python
df2=df1.dropna(axis=0, subset=['Borough'])
df2
```

```python
df3 = df2.groupby(['Postcode', 'Borough'])['Neighbourhood'].agg([('Neighbourhood', ', '.join)]).reset_index()
df3
```

```python
df3.shape
```

Explore and Cluster Neighborhoods

Create Coordinate Dataframe

```python
Coord_df = pd.read_csv('https://cocl.us/Geospatial_data')
Coord_df
```

```python
tor_df = pd.concat([df3, Coord_df], axis=1)
tor_df
```

```python
tor_df2 = tor_df.drop(columns='Postal Code')
tor_df2
```

Map Clusters

```python
import json # library to handle JSON files

!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="tor_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map for Borough and Postcodes
for lat, lng, borough, postcode in zip(tor_df2['Latitude'], tor_df2['Longitude'], tor_df2['Borough'], tor_df2['Postcode']):
    label = '{}, {}'.format(borough, postcode )
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
